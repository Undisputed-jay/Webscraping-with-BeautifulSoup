We will be extracting 25 pages of this website 'https://www.justia.com/lawyers/' to create a dataset of Lawyers from San Francisco, USA. This is for learning purposes and no harm to damage any website.

```python
from bs4 import BeautifulSoup
from IPython.display import IFrame
import requests
import pandas as pd
```

```python
website = 'https://www.justia.com/lawyers/california/san-francisco'
```

```python
name_all = []
city_all = []
phone_all = []
specialization_all = []
address_all = []
university_all = []
email_link_all = []

for y in result:
    try:
        name_all.append(y.find('a', {'class' : 'url main-profile-link'}).get_text().strip())
    except:
        name_all.append('')
    try:
        city_all.append(y.find('span' , {'class' : '-hide-landscape-phone'}).get_text())
    except:
        city_all.append('')
    try:
        phone_all.append(y.find('a' , {'data-vars-action' : 'OrganicListingPhone'}).get('href').split('tel:')[1])
    except:
        phone_all.append('')
    try:
        specialization_all.append(y.find('span' , {'class' : '-practices'}).get_text())
    except:
        specialization_all.append('')
    try:
        address_all.append(y.find('span' , {'class' : '-address -hide-landscape-tablet'}).get_text().strip().replace('\n', '').replace('\t', ''))
    except:
        address_all.append('')
    try:
        university_all.append(y.find('span' , {'class' : '-law-schools'}).get_text())
    except:
        university_all.append('')
    try:
        email_link_all.append(y.find('a', {'class' : '-email'}).get('href'))
    except:
        email_link_all.append('')
```

### Scanning Multiple Pages

```python
name_all2 = []
city_all2 = []
phone_all2 = []
specialization_all2 = []
address_all2 = []
university_all2 = []
email_link_all2 = []

for i in range(1,26):
    mult = 'https://www.justia.com/lawyers/california/san-francisco?page=' + str(i)
    
    response_ = requests.get(mult)
    
    soup_ = BeautifulSoup(response_.content, 'html.parser')
    
    result_ = soup_.find_all('div', {'data-vars-action' : 'OrganicListing'})
    
    for r in result_:
        try:
            name_all2.append(r.find('a', {'class' : 'url main-profile-link'}).get_text().strip())
        except:
            name_all2.append('')
        try:
            city_all2.append(r.find('span' , {'class' : '-hide-landscape-phone'}).get_text())
        except:
            city_all2.append('')
        try:
            phone_all2.append(r.find('a' , {'data-vars-action' : 'OrganicListingPhone'}).get('href').split('tel:')[1])
        except:
            phone_all2.append('')
        try:
            specialization_all2.append(r.find('span' , {'class' : '-practices'}).get_text())
        except:
            specialization_all2.append('')
        try:
            address_all2.append(r.find('span' , {'class' : '-address -hide-landscape-tablet'}).get_text().strip().replace('\n', '').replace('\t', ''))
        except:
            address_all2.append('')
        try:
            university_all2.append(r.find('span' , {'class' : '-law-schools'}).get_text())
        except:
            university_all2.append('')
        try:
            email_link_all2.append(r.find('a', {'class' : '-email'}).get('href'))
        except:
            email_link_all2.append('')
            
        output = {'Lawyer' : name_all2, 'City' : city_all2, 'Phone' : phone_all2, 'Specialization' : specialization_all2, 'University' : university_all2, 'Email Contact' : email_link_all2, 'Address' : address_all2}
        df_ = pd.DataFrame(output)
```
